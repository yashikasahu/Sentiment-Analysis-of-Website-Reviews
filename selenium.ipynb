```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

def scrape_flipkart_reviews(url, num_pages=3):
    options = Options()
    options.add_argument("--start-maximized")
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_argument("user-agent=Mozilla/5.0")

    driver = webdriver.Chrome(options=options)
    driver.get(url)

    # Close login popup if present
    try:
        close_btn = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(),'âœ•')]"))
        )
        close_btn.click()
    except:
        pass

    all_reviews = []

    def get_first_review_text():
        WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "ZmyHeo"))
        )
        reviews = driver.find_elements(By.CLASS_NAME, "ZmyHeo")
        return reviews[0].text.strip() if reviews else None

    for page in range(num_pages):
        # Wait for reviews and scrape
        WebDriverWait(driver, 15).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "ZmyHeo"))
        )
        reviews = driver.find_elements(By.CLASS_NAME, "ZmyHeo")
        print(f"Page {page+1} reviews count: {len(reviews)}")

        for r in reviews:
            all_reviews.append(r.text.strip())

        old_first_review = get_first_review_text()

        # Try to click Next and wait for page to change
        try:
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CLASS_NAME, "_9QVEpD"))
            )
            next_button.click()

            WebDriverWait(driver, 15).until(
                lambda d: get_first_review_text() != old_first_review
            )
        except Exception as e:
            print(f"No more pages or can't navigate further: {e}")
            break

    driver.quit()

    df = pd.DataFrame(all_reviews, columns=["Review"])
    df.to_csv("flipkart_review.csv", index=False)
    print(f"Scraped total {len(all_reviews)} reviews")
    print("Saved to flipkart_reviews.csv")
    return all_reviews

# Use the product reviews page URL here
url = "https://www.flipkart.com/apple-iphone-16-pro-desert-titanium-256-gb/product-reviews/itm4f25cec0bd003?pid=MOBH4DQFEZFXPGNJ"
scrape_flipkart_reviews(url, num_pages=5)
```

```python
# This is the code we have use to do sentiment analysis using VADER 
# VADER (Valence Aware Dictionary and sEntiment Reasoner) is a lexicon and rule-based sentiment analysis tool that is particularly good at analyzing sentiments expressed in social media, reviews, and other short texts.
# It is part of the NLTK (Natural Language Toolkit) library in Python and is designed to handle sentiments expressed in a variety of ways,
# including emoticons, slang, and abbreviations.
# which is inbulit in NLTK library of python.
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
import nltk

nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()

df = pd.read_csv("flipkart_review.csv")

def vader_sentiment(text):
    scores = sid.polarity_scores(text)
    compound = scores['compound']
    if compound >= 0.05:
        return 'POSITIVE'
    elif compound <= -0.05:
        return 'NEGATIVE'
    else:
        return 'NEUTRAL'

df['Sentiment'] = df['Review'].apply(vader_sentiment)

df.to_csv("flipkart_reviews_with_vader_sentiment.csv", index=False)

print(df.head())
```

```python
import pandas as pd
from textblob import TextBlob
from collections import Counter
import re

def generate_review_crux(csv_file):
    # Load reviews
    df = pd.read_csv(csv_file)
    reviews = df["Review"].dropna().tolist()

    # Initialize counters
    positives = []
    negatives = []
    neutral_count = 0

    # Analyze each review
    for review in reviews:
        polarity = TextBlob(review).sentiment.polarity
        if polarity > 0.2:
            positives.append(review)
        elif polarity < -0.2:
            negatives.append(review)
        else:
            neutral_count += 1

    # Word frequency helper
    def get_keywords(texts, stopwords=[], top_n=5):
        words = re.findall(r'\b\w+\b', " ".join(texts).lower())
        filtered = [word for word in words if word not in stopwords and len(word) > 3]
        return [w for w, _ in Counter(filtered).most_common(top_n)]

    # Common themes
    stopwords = set(['phone', 'iphone', 'apple', 'device', 'mobile', 'product', 'very', 'good', 'great', 'nice', 'like', 'love', 'using', 'used'])
    positive_keywords = get_keywords(positives, stopwords)
    negative_keywords = get_keywords(negatives, stopwords)

    # Compose crux
    crux = "Most users praise the iPhone 16 Pro for its "
    crux += ", ".join(positive_keywords[:3]) + ", and overall performance. "
    if negatives:
        crux += "However, some reviewers mentioned issues like "
        crux += ", ".join(negative_keywords[:2]) + ". "
    if neutral_count > 0:
        crux += f"A few reviews were neutral or mixed in sentiment."

    print("\n ----OverAll Review Summary----\n")
    print(crux)

# Example usage
generate_review_crux("flipkart_reviews_with_sentiment.csv")  # Replace with your actual CSV name
```
